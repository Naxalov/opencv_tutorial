```python
import matplotlib.pyplot as plt
import cv2
import numpy as np
```

```python
# utility
def show(img,b=False):
  plt.figure(figsize=(8,8))
  plt.imshow(img,cmap='gray')
  plt.grid(alpha=.2)
  plt.grid(b)
  plt.show()
```

```python
# Read image from file
img = cv2.imread('opencv_logo_solid.jpg')
# Convert image from BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Convert image from RGB to Gray
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
# Show image
plt.imshow(gray, cmap='gray')
plt.show()
```

## Range
The cv2.range is a function that returns an array of evenly spaced numbers over a specified interval. The function has the following syntax:
cv2.inRange(src, lowerb, upperb[, dst])

| Parameter | Condition | Type | Description |
| --- | --- | --- | --- |
| src | - | Mat | input array (single-channel, 8-bit or 32-bit floating point). |
| lowerb | - | Scalar | inclusive lower boundary array or a scalar. |
| upperb | - | Scalar | inclusive upper boundary array or a scalar. |
| dst | - | Mat | output array of the same size as src and CV_8U type. |


```python
mask = cv2.inRange(gray, 180, 182)
show(mask)
```

```python
mask = cv2.inRange(gray, 108,110)
show(mask)
```

```python
img = cv2.imread('banana.jpg')
# Convert BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Convert RGB to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
show(gray)
```

```python
# Calculate the histogram of the image using numpy
histogram = np.histogram(gray, bins=256, range=(100, 250))[0]
print(histogram)
# Plot the histogram
plt.figure()
plt.title("Grayscale Histogram")
plt.xlabel("grayscale value")
plt.ylabel("pixels")
plt.xlim([0.0, 256.0])  # <- named arguments do not work here
plt.plot(histogram)
plt.show()
```

```python
mask = cv2.inRange(gray, 230, 240)
show(mask)
```

```python
lower = np.array([237, 230, 10])
upper = np.array([240, 248, 192])
mask = cv2.inRange(img, lower, upper)
show(mask)
```

```python
# Read image from file
img = cv2.imread('1.png')
# Convert image from BGR to HSV
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
blue = np.array([218, 88, 57])
# Change H, S, V value to OpenCV range
blue[0] = blue[0] // 2
blue[1] = int(blue[1] * 255 / 100)
blue[2] = int(blue[2] * 255 / 100)
print(blue)
lower_blue = np.array([blue[0] - 10,0, 0])
upper_blue = np.array([blue[0] + 10, blue[1] + 30, blue[2] + 30])
mask = cv2.inRange(hsv, lower_blue, upper_blue)
show(mask)

```
