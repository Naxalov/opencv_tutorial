# Introduction to Pixel with OpenCV

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
pixel = np.array([[[255,255,0]]])
# Print the shape of the pixel object
print('Shape of pixel object:', pixel.shape)
plt.imshow(pixel)
plt.show()
```

```python
# Create a pixel with valid range of values which is 0 to 255
# Example for invalid range of values
pixel = np.array([[[256, 256, 256]]])
plt.imshow(pixel)
plt.show()
```
