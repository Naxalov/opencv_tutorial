# Blending images using OpenCV

```python
# Import the necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
# Read the images 
```

```python
# Read the images
logo = cv2.imread('mohirdev_logo.png')
img = cv2.imread('registon_512.png')
# Convert image to rgb
logo = cv2.cvtColor(logo, cv2.COLOR_BGR2RGB)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
# Resize the logo to fit (512, 512)
logo = cv2.resize(logo, (512, 512))
img = cv2.resize(img, (512, 512))
print(logo.shape)
print(img.shape)
```

```python
# Blending the images
# Method 1: Using cv2.addWeighted()
dst = cv2.addWeighted(img, 0.3, logo, 0.6, 50)
plt.imshow(dst)
plt.show()
```
